### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
ingreso_t1 = tienda["Precio"].sum()
ingreso_t2 = tienda2["Precio"].sum()
ingreso_t3 = tienda3["Precio"].sum()
ingreso_t4 = tienda4["Precio"].sum()

print("Ingreso bruto por tienda:")
print(f"Tienda 1: {ingreso_t1:,.0f}")
print(f"Tienda 2: {ingreso_t2:,.0f}")
print(f"Tienda 3: {ingreso_t3:,.0f}")
print(f"Tienda 4: {ingreso_t4:,.0f}")
```

```python
ingreso_real_t1 = (tienda["Precio"] - tienda["Costo de envío"]).sum()
ingreso_real_t2 = (tienda2["Precio"] - tienda2["Costo de envío"]).sum()
ingreso_real_t3 = (tienda3["Precio"] - tienda3["Costo de envío"]).sum()
ingreso_real_t4 = (tienda4["Precio"] - tienda4["Costo de envío"]).sum()

print("\nIngreso neto (descontando costo de envío):")
print(f"Tienda 1: {ingreso_real_t1:,.0f}")
print(f"Tienda 2: {ingreso_real_t2:,.0f}")
print(f"Tienda 3: {ingreso_real_t3:,.0f}")
print(f"Tienda 4: {ingreso_real_t4:,.0f}")
```

# 3. Calificación promedio de la tienda


```python
promedio_t1 = tienda["Calificación"].mean()
promedio_t2 = tienda2["Calificación"].mean()
promedio_t3 = tienda3["Calificación"].mean()
promedio_t4 = tienda4["Calificación"].mean()

print("Calificación promedio por tienda:")
print(f"Tienda 1: {promedio_t1:.2f}")
print(f"Tienda 2: {promedio_t2:.2f}")
print(f"Tienda 3: {promedio_t3:.2f}")
print(f"Tienda 4: {promedio_t4:.2f}")
```

# 4. Productos más y menos vendidos

```python
# Productos más y menos vendidos por tienda

# Tienda 1
ventas_p_t1 = tienda["Producto"].value_counts()
print("Tienda 1 - Más vendidos:")
print(ventas_p_t1.head(5))
print("\nTienda 1 - Menos vendidos:")
print(ventas_p_t1.tail(5))

# Tienda 2
ventas_p_t2 = tienda2["Producto"].value_counts()
print("\nTienda 2 - Más vendidos:")
print(ventas_p_t2.head(5))
print("\nTienda 2 - Menos vendidos:")
print(ventas_p_t2.tail(5))

# Tienda 3
ventas_p_t3 = tienda3["Producto"].value_counts()
print("\nTienda 3 - Más vendidos:")
print(ventas_p_t3.head(5))
print("\nTienda 3 - Menos vendidos:")
print(ventas_p_t3.tail(5))

# Tienda 4
ventas_p_t4 = tienda4["Producto"].value_counts()
print("\nTienda 4 - Más vendidos:")
print(ventas_p_t4.head(5))
print("\nTienda 4 - Menos vendidos:")
print(ventas_p_t4.tail(5))
```

```python
# Productos más vendidos con su categoría vs Categoría mas vendida
for i, df in enumerate([tienda, tienda2, tienda3, tienda4], start=1):
    top_productos = df["Producto"].value_counts().head(5)
    cat_ventas = df.groupby("Categoría del Producto")["Producto"].count().sort_values(ascending=False)
    categoria_top = cat_ventas.index[0]

    print(f"\nTienda {i} - Productos más vendidos con categoría:")
    for producto in top_productos.index:
        categoria_producto = df.loc[df["Producto"] == producto, "Categoría del Producto"].iloc[0]
        pertenece = "✅" if categoria_producto == categoria_top else "❌"
        print(f"{producto} - {categoria_producto} ({pertenece} categoría top: {categoria_top})")
```

# Informe final: Evaluación comparativa de las tiendas de Alura Store LATAM

##Introducción
El objetivo de este análisis fue identificar cuál de las cuatro tiendas de Alura Store LATAM representa la mejor oportunidad de adquisición para el Sr. Juan. Para ello, se evaluaron múltiples variables: ingresos totales y netos, categorías más y menos vendidas, calificaciones promedio de los clientes, productos destacados y costo promedio de envío. Los resultados permiten comprender no solo el rendimiento financiero de cada tienda, sino también su posicionamiento operativo y estratégico dentro del mercado.

##Desarrollo

1. Ingresos y rentabilidad
Las tiendas presentan ingresos netos similares, aunque la Tienda 1 se mantiene como la de mayor facturación (≈ 1.089 millones), seguida de cerca por la Tienda 2 y la Tienda 3. La Tienda 4, en cambio, tiene la menor participación en el total de ingresos (≈ 23.6 %).
Sin embargo, este liderazgo de la Tienda 1 podría estar influenciado por un mayor costo de envío promedio ($26 mil), que encarece la compra final para el cliente y podría afectar la fidelización a largo plazo.

2. Categorías más vendidas
En todas las tiendas, los Muebles y Electrónicos son los segmentos más demandados, lo que demuestra una tendencia clara del mercado hacia productos de alto valor. La Tienda 1 concentra sus ventas en artículos costosos (electrodomésticos, muebles grandes), mientras que la Tienda 4 mantiene una oferta más diversificada, incluyendo productos de ticket medio y bajo (juguetes, artículos para el hogar, libros), lo que podría favorecer un flujo de ventas más constante.

3. Calificación promedio
Las tiendas muestran niveles de satisfacción similares, aunque la Tienda 3 (4.05) y la Tienda 2 (4.04) se posicionan levemente por encima del promedio general, mientras que la Tienda 1 (3.98) presenta la menor valoración. Esto sugiere posibles deficiencias en atención postventa o en la experiencia del cliente, factores cruciales para la sostenibilidad comercial.

4. Productos más y menos vendidos
La Tienda 1 sobresale por su rotación en productos de alto valor unitario (TVs, microondas, armarios), mientras que la Tienda 4 muestra un equilibrio entre categorías, aunque carece de productos “ancla” de alto margen como refrigeradores o guitarras. Este aspecto limita su facturación, pero también reduce su exposición al riesgo de inventarios costosos.

5. Costo promedio de envío
El costo de envío disminuye progresivamente de la Tienda 1 a la 4, siendo la Tienda 4 la más eficiente ($23 mil en promedio). Este factor, aunque no impacta directamente en los ingresos brutos, puede mejorar la competitividad y fidelización de clientes a largo plazo, especialmente si se combina con estrategias de marketing o programas de envío gratuito.

##Conclusión y recomendación

Considerando el balance general entre rentabilidad, diversificación, experiencia del cliente y costos operativos, la mejor opción de compra para el Sr. Juan sería la Tienda 3.
Esta tienda mantiene un equilibrio sólido entre ingresos, calificación de clientes y estructura de costos, sin los extremos de la Tienda 1 (altos ingresos, pero menor satisfacción y costos elevados) ni las debilidades de la Tienda 4 (bajos ingresos pese a buena logística).

No obstante, la Tienda 4 representa una opción estratégica a futuro, especialmente si el objetivo es posicionar una marca con enfoque en accesibilidad y crecimiento sostenido. Su estructura de costos es eficiente y su reputación positiva podría potenciarse mediante la incorporación de productos de mayor valor.
